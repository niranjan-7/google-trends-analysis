```
!pip install pytrends

from pytrends.request import TrendReq

trends = TrendReq(hl='en-US', tz=360)

kw_list = ["Naruto","One Piece","Bleach","Hero Academia","Black Clover"]
trends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```
topic=trends.interest_over_time()
topic.plot()
```
